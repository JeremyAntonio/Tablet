```python
# EJERCICIO 10/2 
void Ordenamiento(){
    int n = 5;  # 1
    int arr[n] = {3, 5, 2, 1, 4};  # 2
    for (int i = 0; i < n; i++){  # 1 + n(1 + .... + 2)
        for (int j = 0; j < n; j++){  # 1 + n(1 + ... + 2)
            if (arr[j] > arr[j + 1]){  # 1 + 1 + (1 + 1) = 4
                aux = arr[j];  # 1 + 1 = 2
                arr[j] = arr[j + 1];  # 1 + 1 + (1 + 1) = 4
                arr[j + 1] = aux;  # (1 + 1) + 1 = 3
            }
        }
    }
}
# 1 + 1 + n(1 + 1 + n(1 + 4 + 2 + 4 + 3 + 2) + 2) = 16n^2 + 4n + 2
```
